# Exploratory Data Analysis (EDA)
## Credit Card Services & Customer Analysis
This notebook explores the financial data to uncover trends in customer spending, demographics, and churn risk.

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')

# Connect to Database
conn = sqlite3.connect('../financial_data.db')

# Load consolidated data
query = """
SELECT 
    c.*, 
    t.Credit_Limit, t.Total_Revolving_Bal, t.Total_Trans_Amt, 
    t.Total_Trans_Vol, t.Avg_Utilization_Ratio, t.Delinquent_Acc
FROM customer c
JOIN credit_card t ON c.Client_Num = t.Client_Num
"""
df = pd.read_sql(query, conn)
df.head()
```

### 1. Demographics Analysis

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(y='Education_Level', data=df, order=df['Education_Level'].value_counts().index, palette='viridis')
plt.title('Distribution of Education Level')

plt.subplot(1, 2, 2)
sns.histplot(df['Income'], kde=True, color='green')
plt.title('Income Distribution')
plt.tight_layout()
plt.show()
```

### 2. Credit Usage vs Income

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Income', y='Credit_Limit', data=df, hue='Card_Owner', alpha=0.6)
plt.title('Income vs Credit Limit')
plt.show()
```

### 3. Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
